```python
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
colors = list()
palette = {0:"red", 1:"green", 2:"blue"}
```

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
for c in np.nditer(iris.target):
    colors.append(palette[int(c)])
```

```python
dataframe = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
scatterplot = pd.plotting.scatter_matrix(dataframe, alpha=0.3, figsize=(10, 10), diagonal='hist', color=colors, marker='o', grid=True)
```
