```python
import pandas as pd
fake_dataset = pd.read_csv('a_loading_example_1.csv', sep=',')
fake_dataset
```

```python
fake_dataset = pd.read_csv('a_loading_example_1.csv', parse_dates=[0])
fake_dataset
```

```python
fake_dataset.fillna(50)  # using 50 as NaN
```

```python
fake_dataset.fillna(-1)
```

```python
fake_dataset.fillna(fake_dataset.mean(axis=0)) # 使用mean方法，计算指定坐标轴上的平均值
# axis=0表示平均值计算是跨行进行的，获得的是列方向的均值
# axis=1表示平均值计算是跨列进行的，获得的是行方向的均值
```

### `fillna`只是以数值的形式填充了缺失值，并没有修改原始数据。如果需要真正的修改它们，需要使用参数`inplace=True`

```python
fake_dataset
```

```python
fake_dataset.fillna(fake_dataset.mean(axis=0), inplace=True)
```
